```python
fig = plt.figure(figsize=(10,5)) # x size y size
plt.suptitle(' ', fontsize = 15, y=0.95)

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.plot(date, high, 'r-')
ax2.plot(date, low)

# ax1 x,y   
ax1.set_title(" ()")
ax1.set_xlabel("") # x  
ax1.set_ylabel("()") # y  
ax1.yaxis.set_tick_params(labelcolor='r') # axis  


ax2.set_title(" ()")
l = ax2.set_xlabel("") # x label   (, )
l.set_color('g')
l.set_fontsize(20)
ax2.set_ylabel("()")

plt.show()
```

###    

-   :anonatate(text, xy, xytext, **kw)
-   : savefig('  ', , dpi)

```python
def annot_max(x,y, ax=None):
    xmax = x[np.argmax(y)]
    ymax = y.max()
    text= "%s, %d"%(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    # text, , 
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.2,0.3), **kw)

def annot_min(x,y, ax=None):
    xmin = x[np.argmin(y)]
    ymin = y.min()
    text= "%s, %d"%(xmin, ymin)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    # text, , 
    ax.annotate(text, xy=(xmin, ymin), xytext=(0.8,0.3), **kw)    
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 5))
plt.suptitle(' ', fontsize = 15, y=1)

xmax = date[np.argmax(high)]
ymax = high.max()

xmin = date[np.argmin(high)]
ymin = high.min()

ax.plot(date, high, 'r-')
plt.plot(xmax, ymax, 'o')
plt.plot(xmin, ymin, 'o')
annot_max(date,high)
annot_min(date,high)

# ax1 x,y   
ax1.set_title(" ()")
ax1.set_xlabel("") # x  
ax1.set_ylabel("()") # y  

plt.savefig('samsung.png', format='png', dpi=300)  #   
plt.show()
```
