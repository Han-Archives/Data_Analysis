### 예제 : 세계 행복 보고서

위 데이터는 [캐글](https://www.kaggle.com/datasets/unsdsn/world-happiness)에서 얻은 데이터로 여러 분야의 전문가들이 155여 국가를 상대로 행복과 관련된 요소를 통해 행복도를 점수로 나타낸 데이터

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import rcParams
rcParams['figure.figsize'] = 13, 7
```

```python
df = pd.read_csv("C:/Users/USER/Desktop/Analysis/Proceed/Kaggle/world_happiness/2019.csv")

df.head()
```

#### 데이터 설명

> Overall rank : 국가 순위

> Country or region : 국가 혹은 도시

> Score (target) : 행복 점수

> GDP per Capita : 1인당 GDP

> Social support : 사회 보장

> Healthy life expectancy : 기대 수명

> Freedom to make life choices : 삶의 대한 선택 자유

> Generosity : 관대함

> Perceptions of corruption : 사회 부패 인식

행복에 대한 기준은 불명확하지만 국가별 행복과 관련된 요소를 찾고 측정했다는 관점에서 데이터를 이해하자.
